# Predicting TV Viewership using Nielsen Ratings and IMDB data

```python
import statsmodels.api as sm
import statsmodels.formula.api as smf
import patsy
import seaborn as sns
from sklearn.linear_model import LinearRegression
from sklearn.linear_model import RidgeCV
import re
import pickle
from operator import itemgetter
import pandas as pd
import numpy as np
from fuzzywuzzy import process
from sklearn.cross_validation import train_test_split
from sklearn import feature_selection as f_select
from sklearn.cross_validation import cross_val_score
from sklearn.cross_validation import KFold
sns.set_style("darkgrid")

%pylab inline
```

## Import, clean and preprocess data

```python
# Load Nielsen data into dataframe

with open("tvseries_data.pkl", 'rb') as picklefile: 
    tvseries_data = pickle.load(picklefile)

df = pd.DataFrame(tvseries_data)
df.head()
```

```python
# Convert Nielsen data to floats and ints where applicable

df['status_year'] = df['status_year'].apply(lambda x: x[:3] + x[-1])
df['status_year'] = df['status_year'].astype(np.int64)

columns = ['demo_yoy', 'viewers_millions', 'viewers_yoy', 'demo_rating']
for c in columns:
    df[c] = df[c].str.strip('%')
    df[c].replace(to_replace="", value=np.nan, inplace=True, )
    df[c].replace(to_replace="n/a", value=np.nan, inplace=True)
    df[c] = df[c].astype(np.float64)
```

```python
#Split title column into titles and status

def clean_title(x):
    if "cancelled" in x:
        a = x.split("(")[0].strip()
        b = "cancelled"
    elif "renewed" in x:
        a = x.split("(")[0].strip()
        b = "renewed"
    else :
        a = x.split("(")[0].strip()
        b = np.nan
    return a, b

df['title'], df['status'] = zip(*df['title'].apply(clean_title))
df['title'] = df['title'].str.lower()
```

```python
# Manual title cleanup

def text_processing(input_str):
    if input_str.find('dwts') >= 0:
        return "dancing with the stars"
    elif input_str.find("afv") >= 0:
        return "america's funniest home videos"
    elif input_str.find("america's fh videos") >= 0:
        return "america's funniest home videos"
    elif input_str.find('fight') >= 0:
        return "the great christmas light fight"
    elif input_str.find('how') >= 0:
        return "how to get away with murder"
    elif input_str.find("shield") >= 0:
        return 'agents of s.h.i.e.l.d'
    elif input_str.find("carter") >= 0:
         return 'agent carter'
    elif input_str.find("paradise") >= 0:
        return 'bachelor in paradise'
    elif input_str.find("ext weight loss") >= 0:
        return 'extreme weight loss'
    elif input_str.find("m. love story") >= 0:
        return 'manhattan love story'
        
    else:
        return input_str


df['title'] = df.title.map(text_processing)
```

```python
#import IMDB data and put in Pandas dataframe
with open("imdb.pkl", 'rb') as picklefile: 
    imdb_data = pickle.load(picklefile)

imdb_df = pd.DataFrame(imdb_data)
imdb_df['title'] = imdb_df['title'].str.lower()
```

```python
# Create list of titles for fuzzy matching
title_names = list(imdb_df['title'])

# Use fuzzy matching to match IMDB and Nielsen Ratings titles for merging purposes
def correct_titles(title):
    if title in title_names:
        return title, 100
    
    new_name, score = process.extractOne(title, title_names)
    if score < 87:
        return title, score
    else:
        return new_name, score

```

```python
df['title'], _ = zip(*df['title'].apply(correct_titles))
```

```python
# Merge IMDB data with Nielsen df

df_merged = pd.merge(df, imdb_df, on='title', how='left')
df_merged.head()
```

```python
#Clear commas from number of votes
df_merged['num_votes'] = df_merged.num_votes.str.replace(',', '')

# Convert imdb data to floats
imdb_cols = ['num_votes', 'premiere_year', 'user_rating']
for col in imdb_cols:
    df_merged[col] = df_merged[col].astype(np.float64)

df_merged.dtypes
```

```python
# Drop premiere year null values, duplicates
df_main = df_merged[df_merged.premiere_year.notnull()].sort_values(by="title")
df_main.drop_duplicates(subset=['status_year', 'title'], inplace=True)

# Impute number of seasons to date
df_main['seasons_to_date'] = df_main['status_year'] - df_main['premiere_year']

# Remove features with lack of data
df_main.drop(['demo_yoy', 'viewers_yoy', 'status'], axis=1, inplace=True)
```

## Linear Regression

```python
lr = LinearRegression()
```

```python
df_main['is_ABC'] = df_main['network'] == 'ABC'
```

```python
y, X = patsy.dmatrices('viewers_millions ~ demo_rating + status_year + C(network)', data=df_main, return_type="dataframe")
```

```python
# Fit linear regression model on full dataset using OLS

model = sm.OLS(y, X)
fit = model.fit()
fit.summary()
```

```python
# Review residuals to ensure homoscedasticity 
fit.resid.plot(style='x', figsize=(8,4));
```

# **Cross-Validation**

```python
# Cross validate results using test/train split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3)
lr.fit(X_train, y_train)
lr.score(X_test, y_test)
```

```python
# Cross validate results using k-folds cross validation
cross_val_score(lr, X, y, cv=5, scoring="r2").mean()
```

## Evaluate Predicted Values

```python
# Predict results using model
df_main['y_predict'] = fit.predict(X)

# View Actual versus Predicted results from model
sns.lmplot('y_predict', 'viewers_millions', hue = 'network', fit_reg=False, data=df_main)
plt.plot(df_main['y_predict'], df_main['y_predict'], color ='darkblue')
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.ylabel('Viewers (in millions)', fontsize=12, weight='bold')
plt.xlabel('Predicted Viewers', fontsize=12, weight='bold')
plt.title("Actual vs Predicted", fontsize=12, weight='bold');
```

```python
# View residual plot of predicted versus actuals 

df_main['resid'] = df_main.y_predict - df_main.viewers_millions

plt.figure(figsize=(5,10))
sns.lmplot('y_predict', 'resid', hue = 'network', fit_reg=False, data=df_main, aspect = 3)
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
plt.ylabel(('Residuals'), fontsize=18, weight='bold')
plt.xlabel('Predicted Viewers', fontsize=18, weight='bold')
plt.title("Residuals", fontsize=18, weight='bold');
```
